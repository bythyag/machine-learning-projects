```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Load the dataset
stationary_series = pd.read_csv('/Users/thyag/Desktop/projects/eloctrocute/sample_data/stationary_battery_soh.csv', index_col=0, parse_dates=True)
```

```python
# Plot ACF and PACF

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ACF Plot
plot_acf(stationary_series, ax=axes[0], lags=40, title='Autocorrelation Function (ACF)')
axes[0].grid(True)

# PACF Plot
plot_pacf(stationary_series, ax=axes[1], lags=40, title='Partial Autocorrelation Function (PACF)')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
